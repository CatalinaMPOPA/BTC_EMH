## CDF 

```python
def PlotCDF(df,title):
    data1 = df['All Data']
    data2 = df['W/out Jumps']
    
    data_sorted1 = np.sort(data1)
    data_sorted2 = np.sort(data2)
    
    p1 = 1. * np.arange(len(data1)) / (len(data1) - 1)
    p2 = 1. * np.arange(len(data2)) / (len(data2) - 1)
    
    fig = plt.figure()
    ax1= fig.add_subplot(1, 1, 1)
    
    ax1.plot(data_sorted1, p1,label='All Data')
    ax1.plot(data_sorted2, p2,label='Without Jumps')
    
    ax1.set_xlabel('$Data$')
    ax1.set_ylabel('$Density$')
    ax1.title.set_text(title)
    
    fig.legend(loc='lower right',  borderaxespad=5)
    fig.show()

    fig.canvas.draw()
```

## CDF- Fit a normal distribution to my data

```python
def PlotCDF3(df, title):  
    data1 = df['All Data']
    data2 = df['W/out Jumps']
    
    p1 = 1. * np.arange(len(data1)) / (len(data1) - 1)
    p2 = 1. * np.arange(len(data2)) / (len(data2) - 1)
    
    fg = plt.figure(); fg.clf()
    ax = fg.add_subplot(1, 1, 1)
    t = np.linspace(0,1, 1000) 
    
    mu1,sigma1 = curve_fit(norm.cdf, data1, p1, p0=[0,1])[0]
    ax.plot(t, norm.cdf(t, mu1, sigma1), alpha=.5)

    mu1,sigma1 = curve_fit(norm.cdf, data2,  p2, p0=[0,1])[0]
    ax.plot(t, norm.cdf(t, mu1, sigma1), alpha=.5)
    
    ax.plot(data1, p1, 'bo',label='All data')
    ax.plot(data2, p2, 'ro', label='Without jumps')
    ax.title.set_text(title)

    fg.legend(loc='lower right',  borderaxespad=3)
   
```

```python
def Plot(df):
    for i in range (0,len(df.sheet_names)):
     PlotPDF3(df.parse(i),df.sheet_names[i])
    print('-------------------------------------')  
```
